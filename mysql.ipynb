CLASSICMODELS.PRODUCT 테이블에서 BUYPRICE 칼럼으로 순위를 매겨 주세요.(오름차순)(ROW_NUMBER,RANK,DENCE RANK 활용)

    select BUYPRICE,
    ROW_NUMBER() OVER (ORDER BY BUYPRICE)  AS ROWNUMBER,
    dense_rank() OVER (order by BUYPRICE) AS DENSERANK, 
    rank() OVER (order by BUYPRICE)  AS RAK
    FROM classicmodels.products;
